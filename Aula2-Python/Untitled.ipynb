```python
import yfinance
```

```python
codigo = input("Digite o código da ação desejada: ")

dados = yfinance.Ticker(codigo).history("6mo")

fechamento = dados.Close

fechamento.plot()
```

### Passo 2

- A análise dos últimos 6 meses
- Cotação máxima
- Cotação minima
- Cotação atual

```python
cotacaoMax = fechamento.max()
cotacaoMin = fechamento.min()
cotacaoAtual = fechamento.iloc[-1]
```

### Passo 3

- abrir uma nova aba (CTRL + T)
- escrever o endereço do gmail
- escrever email e preencher o destinatário
- escrever o assunto
- preencher o corpo do email
- enviar email

```python
!pip install pyautogui
!pip install pyperclip
```

```python
import pyautogui
import pyperclip
```

```python
# configurando passos
pyautogui.PAUSE = 3

# abrir uma nova aba (CTRL + T)
pyautogui.hotkey("ctrl", "t")

# escrever o endereço do gmail
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("enter")

# clicar para escrever o email
pyautogui.click(x=95, y=179)

# preencher o destinatário
pyperclip.copy("sabrinadhedler@gmail.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# preencher o assunto
pyperclip.copy("Análises diárias")
pyautogui.hotkey("ctrl", "v")
pyautogui.hotkey("tab")

# preencher o corpo do email
mensagem = f"""
Prezado gestor,
segue as análises dos últimos seis meses da ação {codigo} conforme solicitado:
Cotação máxima: R$ {round(cotacaoMax, 2)}
Cotação mínima: R$ {round(cotacaoMin, 2)}
Cotação atual: R$ {round(cotacaoAtual, 2)}

Qualquer dúvida estou á disposição!
Atte.
"""

pyperclip.copy(mensagem)
pyautogui.hotkey("ctrl", "v")

# enviar o email
pyautogui.click(x=846, y=694)
```
